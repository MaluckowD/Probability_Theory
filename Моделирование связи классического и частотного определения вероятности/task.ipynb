### Постановка задачи

Рассмотрим алфавит из пяти равновероятных символов:

A, B, C, D, E

Случайным образом генерируется строка длины N, где каждый символ выбирается
независимо и равновероятно.

Событие A — появление символа 'A'.

Цель: показать, что при увеличении числа испытаний частотная вероятность
появления символа 'A' стремится к классической вероятности.


### Классическая вероятность

Так как все символы равновероятны, классическая вероятность появления символа A:

$$P(A) = \frac{1}{5} = 0,2$$


```python
import random
import matplotlib.pyplot as plt
```

```python
alphabet = ['A', 'B', 'C', 'D', 'E']

def random_letter():
    return random.choice(alphabet)
```

```python
N_values = range(10, 5001, 50)
freq_probs = []

for N in N_values:
    count_A = 0
    for _ in range(N):
        if random_letter() == 'A':
            count_A += 1
    freq_probs.append(count_A / N)
```

```python
classic_prob = 1 / 5
classic_probs = [classic_prob] * len(N_values)
```

```python
plt.figure(figsize=(10, 5))

plt.plot(N_values, freq_probs, label='Частотная вероятность')
plt.plot(
    N_values,
    classic_probs,
    linestyle='--',
    label='Классическая вероятность'
)

plt.xlabel('Число испытаний N')
plt.ylabel('Вероятность')
plt.title('Сходимость частотной вероятности к классической')
plt.legend()
plt.grid(True)

plt.show()
```

### Вывод

На графике видно, что при малом числе испытаний частотная вероятность
существенно отклоняется от классической.

Однако с ростом числа испытаний частотная вероятность стремится
к значению 0.2, что подтверждает связь частотного и классического
определений вероятности и иллюстрирует закон больших чисел.
