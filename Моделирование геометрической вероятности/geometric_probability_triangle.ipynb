# Геометрическая вероятность — метод Монте-Карло

## Задача
В квадрате со стороной $a$ расположен равнобедренный треугольник.
Случайная точка равновероятно выбирается в квадрате.

**Найти вероятность попадания точки в треугольник.**

## Теоретическое решение

Площадь квадрата:
$$S_{кв} = a^2$$

Площадь треугольника:
$$S_{тр} = \frac{1}{2}a^2$$

Геометрическая вероятность:
$$P = \frac{S_{тр}}{S_{кв}} = \frac{1}{2}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon
```

```python
a = 10
N = 100

triangle = np.array([
    (0, 0),
    (a, 0),
    (a/2, a)
])

x = np.random.uniform(0, a, N)
y = np.random.uniform(0, a, N)

inside = y <= -2*np.abs(x - a/2) + a
hits = inside.sum()

P_exp = hits / N
P_theory = 0.5

P_theory, P_exp
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

ax.add_patch(Rectangle((0, 0), a, a, fill=False))
ax.add_patch(Polygon(triangle, color='lightblue', alpha=0.5))

ax.scatter(x[inside], y[inside], s=8)
ax.scatter(x[~inside], y[~inside], s=8)

ax.set_aspect('equal')
ax.set_title('Геометрическая вероятность: точка в треугольнике')
plt.show()
```

Экспериментальная вероятность приближается к теоретической при увеличении числа испытаний.
